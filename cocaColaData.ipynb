```python
# Calculating Cost Of Good Sold(COGS)

df['COGS'] = df["Net operating revenues($)"] - df["Operating income"]
df
```

```python
# Calculating the Gross Profit Margin

df["Gross Profit Margin"] = (df["Net operating revenues($)"] - df["COGS"])/df["Net operating revenues($)"]
df
```

```python
# Calculating Net Profit Margin

df['Net Profit Margin'] = (df['Operating income']/df['Net operating revenues($)'])*100
df
```

```python
# Calculating The Return Of Assets(ROA)

df['ROA'] = df['Net operating revenues($)']/df['Total assets($)']
df
```

```python
# Calculating The Return Of Equilty(ROE)

df['ROE'] = df['Operating income']/df['Net income attributable to shareowners of the Coca-Cola Company']
df
```

```python
# Calculating The Debt Coverage

df['Debt Coverage'] = df['Operating income']/df['Long term debt']
df
```

```python
'''Basic and fully diluted shares are how the amount of shares investors hold in a company are measured. 
Basic shares include the stock held by all shareholders, while fully diluted shares are the total number of shares 
if the convertible securities of a company were exercised'''


df['Difference in Shares Income'] = df['Basic net income($)'] - df['Diluted net income']
df
```

From the table, there is very minute difference between the basic net income and diluted net income, this means that the large percentage of the company's securities are convertible shares


```python
# Calculating Retention ratio

df['Retention ratio(%)'] = (df['Basic net income($)'] - df['Cash dividends'])/df['Basic net income($)']*100
df
```

```python
# Yearly trend of Return Of Assets(ROA)

plt.bar(df['Year Ended December 31(In millions except per share data)'], df['ROA'], label='ROA')
plt.xlabel('Year Ended December 31(In millions except per share data)')
plt.ylabel('Return Of Assets')
plt.title('Yearly Trend Of Return Of Assets')
plt.grid(False)
plt.legend()
```

'''The term return on assets (ROA) refers to a financial ratio that indicates how profitable a 
company is in relation to its total assets. ROA can be used by management, analysts, and 
investors to determine whether a company uses its assets efficiently to generate a profit. 
A ROA that rises over time indicates the company is doing well at increasing its profits with each investment dollar it spends.
A falling ROA indicates the company might have over-invested in assets that have failed to produce revenue growth,
a sign the company may be in some trouble. 
ROA can also be used to make apples-to-apples comparisons across companies in the same sector or industry.
'''

Therefore, with the above ROA, we can infer that Coca-cola is a viable company for investment

```python
# Yearly trend of Return Of Equilty

plt.bar(df['Year Ended December 31(In millions except per share data)'], df['ROE'], label='ROE')
plt.xlabel('Year Ended December 31(In millions except per share data)')
plt.ylabel('Return Of Equilty')
plt.title('Yearly Trend Of Return Of Equilty')
plt.grid(False)
plt.legend()
```

Return on equity (ROE) is a measure of financial performance calculated by dividing net income by shareholders' equity. 
Because shareholders' equity is equal to a company’s assets minus its debt, ROE is considered the return on net assets. 
ROE is considered a gauge of a corporation's profitability and how efficient it is in generating profits.
Return on equity (ROE) is the measure of a company's net income divided by its shareholders' equity.
ROE is a gauge of a corporation's profitability and how efficiently it generates those profits.
ROEs will vary based on the industry the company operates in.

```python
# Yearly trend of Debt Coverage

plt.plot(df['Year Ended December 31(In millions except per share data)'], df['Debt Coverage'], label='Debt Coverage')
plt.xlabel('Year Ended December 31(In millions except per share data)')
plt.ylabel('Debt Coverage')
plt.title('Yearly Trend Of Debt Coverage')
plt.grid(False)
plt.legend()
```

The formula for debt coverage ratio is net operating income divided by debt service. The debt coverage ratio is used in banking to determine a companies ability to generate enough income in its operations to cover the expense of a debt.
